#### here we are going to use various ways to draw graphs. Internally, it will use list or numpy and matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt

font = {'family' : 'normal',
        #'weight' : 'bold',
        'size'   : 8}
```

#### bar plot and hist plot
- bar plot is used to draw x vs y
- hist plot is used mainly to view the distribution of your data. So only one variable will be involved. However, you could draw all variables inside an numpy.array to view them together. The data are grouped by 'bins'

```python
# set the figure size
fig = plt.figure(figsize=(5,2))

from sklearn.datasets import load_iris

X = load_iris().data
y = load_iris().target
print(type(X))

# need to get the list out as the first data because bar only takes list
x1 = X[0:1,:10][0]
y1 = [i for i in range(len(x1))]
print (x1)
print (y1)
plt.bar(x1, y1, label='bar chart')
plt.rc('font', **font)

plt.show()

# for hist graph
bins = [1,2,3,4,5,6,8,9,10]
#print("X values are", X[2:5, 0:1])
fig = plt.figure(figsize=(5,3))

plt.hist(X, bins=bins, label='histogram')

# for all graph, you should be able to set all parameters, such as
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title("my python plotting exercise")
plt.legend()
#plt.xlim(0,5)
#plt.ylim(1,3)
plt.rc('font', **font)

plt.show
```

#### Scatter Plot
- the scatter plot is usually used to plot the relationship between two variables. However, it can only plot one comparison at one graph
- to draw multiple ones, seaborn could be used!

```python
fig = plt.figure(figsize=(5,2))
plt.scatter(X[:,0:1], X[:,1:2], label='scatter', s=25, marker='*', color='r')
plt.rc('font', **font)
plt.show()

# seaborn pairplot only takes pd.DataFrame as input
import pandas as pd
import seaborn as sns
df = pd.DataFrame(X, columns=['A', 'B', 'C', 'D'])
fig = plt.figure()
sns.pairplot(df)
plt.rc('font', **font)
plt.show()
```

#### stack plot
- it is used to show the portion of each element inside a whole thing
#### pie chart
- it can also show the percentage of each individual among all

```python
days     = [1,2,3,4,5]
sleeping = [7,8,5,9,6]
eating   = [2,3,4,1,2]
working  = [8,5,4,9,6]
playing  = [7,8,11,5,10]

# as the stack plot doesn't allow you to specify the label, you have to draw
# them in the background
fig = plt.figure(figsize=(6,2))

plt.plot([], [], color='b', label='Sleeping', linewidth=5)
plt.plot([], [], color='r', label='Sleeping', linewidth=5)
plt.plot([], [], color='y', label='Sleeping', linewidth=5)
plt.plot([], [], color='g', label='Sleeping', linewidth=5)

plt.stackplot(days, sleeping, eating, working, playing, colors=['b','r','y','lightgreen'])
plt.legend()
plt.rc('font', **font)
plt.show()

# for pie chart
slices = [7, 2, 2, 13]   # adding everything together equals one day (ie, 24 h)
activities = ['sleeping', 'eating', 'working', 'playing']
colors = ['y', 'g', 'lightblue', 'm']

# try to play each parameter to see the effect
fig = plt.figure(figsize=(3,2))

plt.pie(slices, labels=activities, colors=colors, startangle=90, 
        shadow=True, explode=(0, 0.1, 0, 0), autopct='%1.1f%%')
plt.rc('font', **font)
plt.show()
```

### Figure and Figure parameters
- if you need to modify anything on the graph, you need to reference the default 'figure' first.
- figure and subplot are closely related. Therefore, I will use subplot here as well. But the detailed subplot configuration will be discussed later!

```python
# by default, there is a figure in the background. You need to reference it 
# before you can modify anything on the figure, such as rotating text label
# on the X-axis, or change the size of the figure
fig = plt.figure(figsize=(8,3))
ax1  = plt.subplot2grid((1, 1), (0, 0))

from sklearn.datasets import load_iris
X = load_iris().data
y = load_iris().target

# when you plot, you specify which subplot to draw
ax1.scatter(X[:, 0:1], X[:, 1:2])

# now modify the x-axis labels for each individual subplot
for label in ax1.xaxis.get_ticklabels():
    label.set_rotation(45)
    label.set_fontsize(8)   # this will change x-tick's font size

for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(8)
    
# setting subplot grid
ax1.grid(True, color='y', linestyle='-', linewidth=0.4)
ax1.xaxis.label.set_color('c')
ax1.yaxis.label.set_color('g')
ax1.xaxis.label.set_size(9)   # this will change the x label's font size
ax1.yaxis.label.set_size(9)
ax1.set_xticks([4,5,6,7,8,9])
ax1.tick_params(axis='x', colors='#f06215')

ax1.spines['left'].set_color('c')
ax1.spines['left'].set_linewidth(1)
#ax1.spines['right'].set_visible(False)
#ax1.spines['top'].set_visible(False)
#ax1.spines['bottom'].set_visible(True)
ax1.spines['bottom'].set_color('b')
ax1.spines['bottom'].set_linewidth(1)

# to adjust subplot more
#  #the last two are padding for multiple subplots
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.6, top=0.5, 
                    wspace=0.1, hspace=0.2) 
plt.rc('font', **font)
plt.xlabel("pedal width")
plt.ylabel("pedal length")
plt.show()
```

### Subplots
in order to make subplots , you need to explicitly specify it as well. There are 4 major ways to make subplots
- ax1.add_subplot
- plt.subplot2grid
- gridspect
- plt.subplots

```python
# first approach: using add_subplot()
style.use('fivethirtyeight')
X = load_iris().data

fig = plt.figure()

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(2,1,2)

ax1.scatter(X[:,0:1], X[:,1:2])
ax2.scatter(X[:,0:1], X[:,2:3])
ax3.scatter(X[:,2:3], X[:,1:2])

ax1.tick_params(axis='x', labelsize=8)
ax1.tick_params(axis='y', labelsize=8)
ax2.tick_params(axis='x', labelsize=8)
ax2.tick_params(axis='y', labelsize=8)
ax3.tick_params(axis='x', labelsize=8)
ax3.tick_params(axis='y', labelsize=8)

plt.subplots_adjust(wspace=0.3, hspace=0.3) 

plt.show()
```

#### subplot 2

```python
# second approach: use subplot2grid()
style.use('ggplot')
X = load_iris().data

fig = plt.figure()

# the first tuple is the shape of the subplot, such as (1, 1) means 
# there is only one subplot that is ==> 1 by 1.
# the second tuple is the starting point of the subplot
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((6,1), (1,0), rowspan=3, colspan=1)
ax3 = plt.subplot2grid((6,1), (4,0), rowspan=2, colspan=1)

ax1.scatter(X[:,0:1], X[:,1:2])
ax2.scatter(X[:,0:1], X[:,2:3])
ax3.scatter(X[:,2:3], X[:,1:2])

ax1.tick_params(axis='x', labelsize=8)
ax1.tick_params(axis='y', labelsize=5)
ax2.tick_params(axis='x', labelsize=8)
ax2.tick_params(axis='y', labelsize=8)
ax3.tick_params(axis='x', labelsize=8)
ax3.tick_params(axis='y', labelsize=8)

plt.subplots_adjust(wspace=0.3, hspace=0.6) 

plt.show()
```

```python
from matplotlib.gridspec import GridSpec
plt.figure()

gs = GridSpec(3,3) # nrows = 3 and ncols = 3
ax1 = plt.subplot(gs[0,:])
ax2 = plt.subplot(gs[1,:2])
ax3 = plt.subplot(gs[1:,2])
ax4 = plt.subplot(gs[-1,0])
ax5 = plt.subplot(gs[-1,-2])

ax1.scatter(X[:,0:1], X[:,1:2])
ax2.scatter(X[:,0:1], X[:,2:3])
ax3.scatter(X[:,2:3], X[:,1:2])
ax4.scatter(X[:,0:1], X[:,3:])
ax5.scatter(X[:,2:3], X[:,3:])

plt.tight_layout()
plt.show()
```

#### subplot 4
- plt.subplots() is a function that returns a tuple containing a figure and axes object(s). Thus when using fig, ax = plt.subplots() you unpack this tuple into the variables fig and ax. 
- Having fig is useful if you want to change figure-level attributes or save the figure as an image file later (e.g. with fig.savefig('yourfilename.png'). 
- Also, all axes objects (the objects that have plotting methods), have a parent figure object anyway, thus: fig, ax = plt.subplots() is equal to:
  - fig = plt.figure()
  - ax = fig.add_subplot(1, 1, 1)

```python
# using subplots()
fig, ((ax11, ax12), (ax21, ax22)) = plt.subplots(2, 2, sharex=True, sharey=True)

ax11.scatter(X[:,0:1], X[:,1:2])
ax12.scatter(X[:,0:1], X[:,2:3])
ax21.scatter(X[:,2:3], X[:,1:2])
ax22.scatter(X[:,0:1], X[:,3:])

plt.tight_layout()
plt.show()

nrows = 2
ncols = 3

for i in range(nrows):
    fig, axes = plt.subplots(nrows=1, ncols=ncols, figsize=(8, 3), sharey=True)
    for j in range(ncols):
        a = j+1
        b = j+2
        axes[j].scatter(X[:,j:a], X[:,a:b])
        
plt.tight_layout()
plt.show()
```

#### use seaborn library for plotting
